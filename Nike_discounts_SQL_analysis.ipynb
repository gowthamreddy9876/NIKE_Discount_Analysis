#avg_discounts

```python
%sql
SELECT AVG(discount_percent) AS average_discount_percent
FROM workspace.nike_discounts.nike_discounts_clean;
```

#Highest_discount_product


```python
%sql
SELECT 
  title,
  product_type,
  original_price,
  discount_percent,
  final_price
FROM workspace.nike_discounts.nike_discounts_clean
ORDER BY discount_percent DESC
LIMIT 10;
```

#Discount Price

```python
%sql
SELECT
  discount_percent,
  COUNT(*) AS discount_count
FROM workspace.nike_discounts.nike_discounts_clean
GROUP BY discount_percent
ORDER BY discount_percent;

```

# TO use counters in Dashboard

```python
%sql
SELECT
  AVG(original_price) AS avg_original_price,
  AVG(final_price) AS avg_final_price,
  AVG(original_price - final_price) AS avg_discount_amount
FROM workspace.nike_discounts.nike_discounts_clean;
```

# Highest final price products

```python
%sql
SELECT
  title,
  final_price
FROM workspace.nike_discounts.nike_discounts_clean
ORDER BY final_price DESC
LIMIT 15;
```

#Discount_amount_range

```python
%sql
SELECT
  CASE 
      WHEN discount_amount BETWEEN 0 AND 20 THEN '0–20'
      WHEN discount_amount BETWEEN 21 AND 40 THEN '21–40'
      WHEN discount_amount BETWEEN 41 AND 60 THEN '41–60'
      WHEN discount_amount BETWEEN 61 AND 80 THEN '61–80'
      WHEN discount_amount BETWEEN 81 AND 100 THEN '81–100'
      ELSE '100+'
  END AS discount_amount_range,
  COUNT(*) AS num_products
FROM workspace.nike_discounts.nike_discounts_clean
GROUP BY discount_amount_range
ORDER BY num_products DESC;
```
